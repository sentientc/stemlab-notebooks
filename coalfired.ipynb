Health risk of Coal-fired power plant
-------------
1. In addition to particulate matter(PM) emission, coal-fired power plant may emit a number of pollutants that is considered health risk(EH&E, 2011).
2. non-carcinogenic health risk from SO2 and Hg
3. carcinogenic health from As and Cr
4. Seleniosis risk from Se
5. Quality of coal affect emissions. 
6. Health risk(non-carcinogenic) of PM is evaluated in terms of mortality(Penney, Bell, & Balbus, 2009).
7. PM2.5 dose response: 10 Î¼g/m3 increase in ambient PM2.5 concentration was associated with approximately a 4% increased risk of all-cause mortality. A 6% increased risk of cardiopulmonary mortality and a 8% increased risk of lung cancer mortality for individuals 30 and older.

6\. Mokhtar, M. M., Hassim, M. H., & Taib, R.M. (2014). Health risk assessment of emissions from a coal-fired power plant using AERMOD modelling. Process Safety and Environmental Protection. 92: 476â485  
  + coal-fired power plants emit SOx, NOx, carbon dioxide (CO2), hydrogen fluoride (HF), hydrogen chloride (HCl), mercury (Hg) and dioxins. This health risk assessment(HRA) assess non-carcinogenic health risk from SO2 and Hg and carcinogenic health from As and Cr.

7\. Bolten, J. G. et. al.(1987). Health risk of toxic emissions from a coal-fired power plant. Electric Power Research Institute.  
  + Assess chronic health risk of coal-fired power plant toxic emissions.
  + Arsenic, beryllium and selenium concentrations in local rivers.
  + Concentration analysis using surface water transport model.
  + Exposure model utilise pathways from surface emission to consumption of water and fish from the affected waterways and consumption of milk, beef, pork, paultry, grain and vegetables produced from the affected water irrigated land.

8\. Jozi, S. A. & Pouriyeh, A. A. (2011). Health-safety and environmental risk assessment of power plants using multi criteria decision making method. Association of the Chemical Engineers of Serbia.
  + Assess environmental risk using emissions(CO2, CO, NOx ,and SO2), effluents(BOD, COD, DO, TDS, TSS and other pollutants), noise and vibration.
  + Assess risk with analytical hierarchy process(AHP).  hierarchical structure of the risk is a variety of sub-criteria including health-safety risks, technological risks, and the affected environment risks. Each sub-criterion can further divided into the sub-sub-criteria. For example, safety and health risks include risks and diseases caused by thermal stress, hearing loss, skeleton and muscular disorders, skin and respiratory diseases, vision loss, accident, electric shock and illness resulting from exposure to magnetic field over a long term.
  + 9 point judgment scale of Safety, one score was assigned to each pair-wise comparison of risks and effects. The Expert Choice software the relative importance of criteria were calculated and the weight of final risk was obtained by aggregating of multiplied criteria in options weight. 

9\. BARCLAYS (2015) Environmental and Social Risk Briefing: power generation. Barclays banks PLC.
  + coal fired power stations the coal is pulverised (to enhance combustion efficiency), then fed into the combustion chamber of a boiler and burned. Advanced coal utilisation technology (e.g. fluidised bed combustion) tends to be more efficient than conventional and combined cycle systems. Integrated Coal Gasification Combined Cycle (IGCC) power plant is the most environmentally friendly coal-fired power generation technology. Most importantly, coal gasification offers the immediate opportunity to generate power with near zero greenhouse gas emissions and the pathway to a future hydrogen economy.
  + There are limited alternatives to coal, technology should be installed to minimise emissions as much as possible.
