# Atmospheric dispersion

### Guassian dispersion model(wikipedia)
$$C = \frac{Q}{u} \cdot \frac{f}{\sigma_y \sqrt{2 \pi}} \cdot \frac{g}{\sigma_z\sqrt{2 \pi}}$$

$C$ concentration of emissions, in $g/m^3$, at any receptor located at x, y, and z  
  - $x$ meters downwind from the emission source point  
  - $y$ meters crosswind from the emission plume centerline  
  - $z$ meters above ground level  
  
$\sigma _{z}$ vertical standard deviation of the emission distribution, in m  
$\sigma _{y}$ horizontal standard deviation of the emission distribution, in m  
$f$ crosswind dispersion parameter where $f=exp[-y^2/(2\sigma_y^2)]$  
$H$ height of emission plume centerline above ground level, in m  
$Q$ source pollutant emission rate, in g/s  
$g$ vertical dispersion parameter  
  - $g=g_1+g_2+g_3$  
  - $g_1$ vertical dispersion with no reflections; $\exp [-(z-H)^{2}/(2\sigma _{z}^{2})]$  
  - $g_2$ vertical dispersion for reflection from the ground; $\exp [-(z+H)^{2}/(2\sigma _{z}^{2})]$  
  - $g_3$ vertical dispersion for reflection from an inversion aloft(?)  

$\sigma _{z}$ and $\sigma _{y}$ are functions of the atmospheric stability class (i.e., a measure of the turbulence in the ambient atmosphere) and of the downwind distance to the receptor. The two most important variables affecting the degree of pollutant emission dispersion obtained are the height of the emission source point and the degree of atmospheric turbulence. The more turbulence, the better the degree of dispersion.  
  - $\sigma_y(x) = exp(I_y + J_y ln(x) + K_y[ln(x)]^2)$    
  - $\sigma_z(x) = exp(I_z + J_z ln(x) + K_z[ln(x)]^2)$  
  - $I_y, J_y, K_y, I_z, J_z, and K_z$ are values in the six stability classes are referred to: A-extremely unstable B-moderately unstable C-slightly unstable D-neutral E-slightly stable F-moderately stable.

##### AERMOD(MFD)
"plume sections are emitted into a traveling train of convective elements - updrafts and downdrafts - that move with the mean wind. The vertical and lateral velocities in each element are assumed to be random variables and characterized by their probability density functions (pdf). The mean concentration is found from the pdf of the position of source-emitted â€œparticles". "  
  - In the CBL, the pdf of the vertical velocity (w) is positively skewed and results in a non-Gaussian vertical concentration distribution Fz  
  - The lateral velocity pdf is approximately Gaussian and this pdf and the resulting concentration distribution, Fy, are assumed to be Gaussian.  
 $F_y=\frac{1}{\sqrt{2\pi} \sigma_y} exp(\frac{-y^2}{2\sigma_y^2})$

"Meander (or the slow lateral back and forth shifting of the plume) decreases the likelihood of seeing a coherent plume after long travel times. This effect on plume
concentration could best be modeled with a particle trajectory model, since these models estimate the concentration at a receptor by counting the number of times a particle is seen in the receptor volume. However, as a simple steady state model, AERMOD is not capable of producing such information. AERMOD accounts for meander by interpolating between two concentration limits: the coherent plume limit (which assumes that the wind direction is distributed about a welldefined mean direction with variations due solely to lateral turbulence) and the random plume limit, (which assumes an equal probability of any wind direction)."

## CMAQ dispersion
diffusion equation
$$ \frac{\partial(\varphi_i J_s)}{\partial(t)}+m^2\nabla_s \cdot (\frac{\varphi_i J_s \hat{V_s}}{m^2}) + \frac{\partial (\varphi_i J_s \dot{s})}{\partial s}=J_sQ_{\varphi_i}$$
  - $\varphi_i$ trace speices concentration  
  - $J_s$ vertical Jacobian of the terrain-influence coordinate s
  - $m$ map scale factor
  - $\hat{V_s}$ vertical wind component  
  - $Q_{\varphi_i}$ source or sink term
